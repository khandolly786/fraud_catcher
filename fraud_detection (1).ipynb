**DATA CLEANING**


*   since there are null values in the dataset and their occurance is only once, thus we can safely remove that row (as deleting one row won't affect the dataset containing over 5 million rows)





```python
df=data.dropna() #removing the null values
```

```python
df.isna().sum()
```

The sender's and receiver's id's are of no use as they don't really contribute in prediction. Thus they are removed from the dataset and new dataset is stored in df variable. Same goes for type of transacion, i tried applying hot label encoding to include it, but it did not gave any better results so i just removed it.

```python
df=df.drop(['nameOrig','nameDest','type'],axis=1)
```

```python
df
```

Now we're assigning our X and Y variables with the values from the dataset.
X includes all the columns if df except isFraud as it our dependent variable.

```python
X=df.loc[:,df.columns!='isFraud']
Y=df['isFraud']
```

```python
X
```

Now we're dividing the dataset into training and testing data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.40,random_state=10)
```

Finally i applied LogisticRegression algorithm (as the x and y graph tend to be in a straight line (linear regression)).

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression()
clf.fit(X_train, Y_train)
Y_pred = clf.predict(X_test)
print ("Accuracy : ", clf.score(X_test, Y_test))
```

**The model was finally able to detect the fraud cases with an accuracy of 99.83 %**
